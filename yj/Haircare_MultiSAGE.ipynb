```python
import dgl
from scipy import spatial
import numpy as np
import pandas as pd

df = pd.read_excel('C:/Users/KYJ2/Desktop/ESG-Recommender-System-by-MultiSAGE/dataset/Haircare/df.xlsx')
merged_df = pd.read_excel('C:/Users/KYJ2/Desktop/ESG-Recommender-System-by-MultiSAGE/dataset/Haircare/merged_df.xlsx')
attributes = pd.read_excel('C:/Users/KYJ2/Desktop/ESG-Recommender-System-by-MultiSAGE/dataset/Haircare/attributes.xlsx')
```

```python
df = df.drop_duplicates(subset=['상품명'])
df
```

```python
df1 = df.loc[:,(df.columns != '제조사') & (df.columns != '브랜드')]
df1
```

```python
small_df = df1.explode('특징')[['ID','상품명','특징']]
small_df
```

```python
attributes = pd.DataFrame(small_df['특징'].unique()).reset_index()
attributes.columns = ['attribute_id', '특징']
attributes = pd.DataFrame(attributes).astype({'attribute_id': 'category'})
attributes
```

```python
merged_df = pd.merge(small_df, attributes, on=['특징'])
```

```python
small_df = small_df[small_df['특징'].isna() == True]
```

```python
from multisage.builder import PandasGraphBuilder

graph_builder = PandasGraphBuilder()
graph_builder.add_entities(attributes, 'attribute_id', '특징')
graph_builder.add_entities(df, 'ID', '상품')
graph_builder.add_binary_relations(merged_df, 'attribute_id', 'ID', 'define')
graph_builder.add_binary_relations(merged_df, 'ID', 'attribute_id', 'define-by')
g = graph_builder.build()
g
```

```python
# 그래프 구조 확인

print("Number of nodes per type:",g.number_of_nodes())
print("Number of edges per type:",g.number_of_edges())
print("Node types:",g.ntypes)
print("Edge types:",g.etypes)

# 노드 데이터 확인
for ntype in g.ntypes:
    print(f"Node type: {ntype}")
    print(g.nodes[ntype].data)

# 엣지 데이터 확인
for stype, etype, dtype in g.canonical_etypes:
    print(f"Edge type: {etype}")
    print(g.edges[etype].data)

# 그래프 구조 확인
print("Graph structure:")
print(g)
```
